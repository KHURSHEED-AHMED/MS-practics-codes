```python
# load the mnist handwritten digits data form the keras datasets
from tensorflow.keras.datasets import mnist
```

```python
#split the training and test along with its labels
(train_images,train_labels),(test_images, test_labels) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
```

```python
# create the fig variable in which give the figure the function
fig = plt.figure()
#iterate the loop 9 time to create the 9 sbuplots with 3x3 grid
for i in range(12):
  plt.subplot(4,3,i+1)
  #tight_layout show the no space btw these sbuplots
  plt.tight_layout()
  plt.imshow(train_images[i],cmap="gray",interpolation="none")
  plt.title("Digit:{}".format(train_labels[i]))
  plt.xticks([])
  plt.yticks([])
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
#create teh model hidden layers with input layer its dimension is 28*28 last layer has 10 neurons bcz it as 10 classes
model = keras.Sequential([
    layers.Dense(512, activation="relu",input_shape=(28*28,)),
    layers.Dense(256,activation="relu"),
    layers.Dense(10,activation="softmax")

])
```

```python
model.summary()
```

```python
#compile the model and track the accuray
model.compile(optimizer="rmsprop", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
#flatten/Dense layer take the input in one dimension so we have to cover 1 dim from 28 * 28 = 784
#normalize the pixal values form 0-255 to 0-1 for faster and more stable training
train_images = train_images.reshape((60000,28*28))
train_images = train_images.astype("float32")/255
test_images = test_images.reshape((10000,28*28))
test_images = test_images.astype("float32")/255
```

```python
#train the model
model.fit(train_images,train_labels,epochs=5,batch_size=64)
```

```python
# test image shape is 1D we have conver it bcz of training by Dense layer
test_images[2555].shape
```

```python
#now we have to conver its dimension for prediction
import numpy as np
img = np.expand_dims(test_images[2555],axis=0)
```

```python
#now it is two dim
img.shape
```

```python
prediction = model.predict(img)
prediction
```

```python
#sum of all the probablitites is equal to 1
prediction[0].sum()
```

```python
#now predict the whole test_images
y_pred = model.predict(test_images)
```

```python
y_pred
```

```python
# model.evaluate(test_images,test_labels)
test_loss,test_acc = model.evaluate(test_images,test_labels)
print(f"test_acc:{test_acc}")
print(f"test_loss:{test_loss}")
```

```python
#argmax is use to pick class with the higest probability for each image
y_preds = np.argmax(y_pred,axis=1)
```

```python
# here are these classes with higest probablities
y_preds
```

```python
# show the performance through the confusion_matrix
from sklearn.metrics import confusion_matrix
result = confusion_matrix(test_labels,y_preds)
result
```
